## Feature Engineering

**Features to be created:**
   - Distance from N-Day Moving Average
   - Distance from N-Day High/Low Rolling features
   - Price Distance from N-Day Lag
    

```python
# FEATURES

data = raw_data.copy()

# Returns
data['ret'] = data['Close'] / data['Close'].shift(1) - 1


# Distance from Moving Averages
for m in [10, 20, 30, 50, 100]:
    data[f'feat_dist_from_ma_{m}'] = data['Close'] / data['Close'].rolling(m).mean() - 1
    

# Distance from n day max/min
for m in [6, 10, 15, 20, 30, 50, 100]:
    data[f'feat_dist_from_max_{m}'] = data['Close'] / data['High'].rolling(m).max() - 1
    
    data[f'feat_dist_from_min_{m}'] = data['Close'] / data['Low'].rolling(m).min() - 1


# Price Distance
for m in [6, 10, 15, 20, 30, 50, 100]:
    data[f'feat_price_dist{m}'] = data['Close'] / data['Close'].shift(m) - 1
    
data
```

## Create the Target Variable

```python
#TARGET VARIABLE

# Target variable (Predict price above 20SMA in 5 bars)
data['target_ma'] = data['Close'].rolling(20).mean()
data['price_above_ma'] = data['Close'] > data['target_ma']
data['target'] = data['price_above_ma'].astype(int).shift(-5)

data
```

```python
# Assuming df is your OHLC DataFrame with columns: Date, Open, High, Low, Close


def profit_targets(data):
    # Step 1: Calculate the 14-period ATR
    #data['ATR'] = data['High'].rolling(window=14).max() - data['Low'].rolling(window=14).min()

    # Step 2: Calculate upper and lower bounds
    data['Take_profit'] = data['Close'] * (1 + 0.02)
    data['Stop_loss'] = data['Close'] * (1 - 0.02)

    # Step 3: Initialize the Indicator column to 0
    data['Indicator'] = 0

    data.dropna(inplace=True)

    # Step 4: Calculate the indicator column (1 if upper bound hit first, -1 if lower bound hit first, 0 otherwise)
    for i in range(len(data)):
        takeprofit_hit = np.where(data['High'].iloc[i+1:i+11] >= data['Take_profit'].iloc[i])[0]
        stoploss_hit = np.where(data['Low'].iloc[i+1:i+11] <= data['Stop_loss'].iloc[i])[0]

    #     print(f'Currently at {i}')

    #     print(takeprofit_hit)
    #     print(len(takeprofit_hit))
    #     print(stoploss_hit)

        # if no target hit.
        if len(takeprofit_hit) == 0:
            if len(stoploss_hit) > 0:
                data.at[i, 'Indicator'] = -1
            else:
                data.at[i, 'Indicator'] = 0
        elif len(stoploss_hit) == 0:
            data.at[i, 'Indicator'] = 1
        elif np.min(takeprofit_hit) <  np.min(stoploss_hit):
            data.at[i, 'Indicator'] = 1
        elif np.min(takeprofit_hit) > np.min(stoploss_hit):
            data.at[i, 'Indicator'] = -1
    
    return data
```

## Train and Test Split

```python
# CLEAN DATA
dataset = data.dropna()
dataset.reset_index(inplace=True,drop=True)
dataset
```

```python
feat_cols = [col for col in dataset.columns if 'feat' in col]

# Training data
TRAIN_UNTIL = '2023-06-30'
train_until = TRAIN_UNTIL

# training_data
X_train = dataset[dataset['Time'] <= train_until][feat_cols]
y_train = dataset[dataset['Time'] <= train_until]['target']

# test_data
X_test = dataset[dataset['Time'] > train_until][feat_cols]
y_test = dataset[dataset['Time'] > train_until]['target']

# returns
ret_test = dataset[dataset['Time'] > train_until]['ret']
```

## Train Models

Below we will train several models and assess their accuracy. These models will include:
 - Random Forest Classifier
 - Gradient Boosting Classifier; and
 - Logistic Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, recall_score, precision_score

def train_logistic_regression(X_train, y_train):
    model = LogisticRegression()
    model.fit(X_train, y_train)
    return model

def train_gradient_boosting(X_train, y_train):
    model = GradientBoostingClassifier(n_estimators=100)
    model.fit(X_train, y_train)
    return model

def train_random_forest(X_train, y_train):
    model = RandomForestClassifier(n_estimators=100)
    model.fit(X_train, y_train)
    return model

def evaluate_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    auc = roc_auc_score(y_test, model.predict_proba(X_test)[:,1])
    recall = recall_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    return accuracy, auc, recall, precision
```

```python
# Train and evaluate Logistic Regression
lr_model = train_logistic_regression(X_train, y_train)
lr_results = evaluate_model(lr_model, X_test, y_test)

# Train and evaluate Gradient Boosting
gb_model = train_gradient_boosting(X_train, y_train)
gb_results = evaluate_model(gb_model, X_test, y_test)

# Train and evaluate Random Forest
rf_model = train_random_forest(X_train=X_train, y_train=y_train)
rf_results = evaluate_model(rf_model, X_test=X_test, y_test=y_test)

# Create a DataFrame with the results
results_df = pd.DataFrame({
    'Model': ['Gradient Boosting', 'Random Forest', 'Logistic Regression'],
    'Accuracy': [gb_results[0], rf_results[0], lr_results[0]],
    'AUC': [gb_results[1], rf_results[1], lr_results[1]],
    'Recall': [gb_results[2], rf_results[2], lr_results[2]],
    'Precision': [gb_results[3], rf_results[3], lr_results[3]]
})

results_df
```

```python
results = pd.DataFrame({'returns':ret_test, 'prediction':y_pred})
results['position'] = np.where(results['prediction'] == 1, -1,1)

results
```

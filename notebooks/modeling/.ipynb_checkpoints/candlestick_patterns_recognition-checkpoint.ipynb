```python
# Import Libraries
import pandas as pd
import numpy as np
from pylab import mpl, plt
import mplfinance as mpf
import seaborn as sns
# plt.style.use('seaborn')
# mpl.rcParams['font.family'] = 'serif'
import talib as ta
from talib import abstract
from scripts.data_download import get_bybit_data
```

```python
# IMPORT DATA
symbol = 'BTCUSDT'
start_time = '2020-01-01 00:00:00'

raw_data = get_bybit_data(product_type='linear', symbol=symbol, interval=60,
                      start_time=start_time,
                      verbose=False)

raw_data.head()
```

```python
# Plot OHLC data
raw = raw_data.set_index('Time')
mpf.plot(raw,
        type='line',
        title = symbol,
        style='yahoo')
```

```python
ret = raw['Close']/ raw['Close'].shift(1) - 1

sns.histplot(ret, kde=True, color='blue', bins=100)
plt.title('Returns Distribution')
```

```python
# all Pattern Recognition indicators start with CDL
all_indicators = [method for method in dir(abstract) if method.startswith('CDL')]
```

```python
# Add SMA to data (For Trend detection)
data = raw.copy()

### The TA package doesnt seem to have a function for the donchain indicator. So will need to create my own.
data['SMA'] = ta.SMA(data['Close'],20)
data['DownTrend'] = np.where(data['Close'] < data['SMA'], True, False)
data.columns = data.columns.str.lower()
data
```

```python
for indicator in all_indicators:
    data[str(indicator)] = getattr(abstract, indicator)(data)
data
```

```python
# Count the number of occurances of each pattern
candlestick_count = data.drop(['open', 'high','low','close','volume','sma','downtrend'], axis=1).abs().sum()/100
print(candlestick_count.sort_values(ascending=False))
type(candlestick_count)
```

```python

positive_count = candlestick_count[candlestick_count > 0]

# Plotting
%matplotlib inline
ax = positive_count.sort_values(ascending=False).plot.bar()
#ax.grid()
ax.set_xlabel('Pattern')
ax.set_ylabel('Count')
ax.set_title('Frequency of Patterns in BTC')
```

```python
# Get the n-period return after pattern occurance

# Calculate 5-period return
data['return_5'] = data['close'].shift(-5) / data['close'] - 1

# Calculate 10-period return
data['return_10'] = data['close'].shift(-10) / data['close'] - 1

# Calculate 15-period return
data['return_15'] = data['close'].shift(-15) / data['close'] - 1
```

```python
# Summarise the distributions of the n-period return for each pattern



# pattern | n-period| mean | stdev | median | 25pct | 75pct |
results = []
for pattern in all_indicators:
    temp = data[data[pattern] > 0]
    stats = temp['return_10'].describe()
    
    result = {'pattern':pattern, 'n_period':10, 'count':stats['count'],
             'mean':stats['mean']}
    
    results.append(result)

results_df = pd.DataFrame(results)
results_df.sort_values('count', ascending=False, inplace=True)
results_df
```

```python
# Add Indicator for Hammer candlestick pattern
data['Hammer'] = ta.CDLHAMMER(data['Open'], data['High'], data['Low'], data['Close'])

data.tail()
```

```python
# Plot OHLC data with markers to indicate Hammer cadlestick pattern
```

```python
# Start Zooming into cases where there is a down trend and hammer pattern
data['Signal'] = np.where(data['DownTrend'] & (data['Hammer']==100) & (data['Close'] > data['Open']), 1, 0)

signals = data.loc[data['Signal'] == 1]

signals['Signal'].count()
```

```python
# Add marker to where the signal is.
data['Signal_Marker'] = np.where(data['Signal'] == 1, data['Signal'] * data['Low'], np.nan)

data.loc[data['Signal']==1]
```

```python
# Plot chart with entry signals

ap = [mpf.make_addplot(data['Signal_Marker'], scatter=True, markersize=50, marker=r'$\Uparrow$', color='green')]
mpf.plot(data, type='candle', addplot=ap, )
```

```python
# Create a list to hold the subplots
subplots = []

# Loop through each signal and plot 50 bars before and after
for signal_index in signals.index[0:20]:
    start_index = signal_index - pd.Timedelta(minutes=50*15)
    end_index = signal_index + pd.Timedelta(minutes=50*15)
    sub_data = data.loc[start_index:end_index].copy()
    ap = [mpf.make_addplot(sub_data['Signal_Marker'], scatter=True, markersize=100, marker=r'$\Uparrow$', color='green')]
    subplot = mpf.plot(sub_data, type='candle', addplot=ap, returnfig=True)
    subplots.append(subplot)

# Show the subplots
for subplot in subplots:
    mpf.show(subplot)
```

```python
# Loop through each signal and plot the cummulative return (50 bars):

fig, ax = plt.subplots()

my_list = []

for signal_index in signals.index:
    
    end_index = signal_index + pd.Timedelta(minutes=50*15)
    sub_data = data.loc[signal_index:end_index].copy()
    sub_data['return'] = np.log(sub_data['Close']/sub_data['Close'].shift(1))
    sub_data['creturns'] = sub_data['return'].cumsum().apply(np.exp)
    
    # append final values:
    my_list.append(sub_data.iloc[-1]['creturns'])
    
    ax.plot(np.array(sub_data['creturns']), linewidth=0.8)
```

```python
plt.hist(my_list, bins=100)
plt.title('Distribution of Final Values')
plt.xlabel('Cummulative return')
plt.ylabel('Frequency')
plt.show()
```
